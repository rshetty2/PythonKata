```python
#9. Drop rows with missing values.
df_cleaned = df.dropna()

print(df_cleaned)
```

```python
#10. Replace missing values in the 'bhk' column with the string 'Unknown'.
df_na_replaced=df.fillna({'bhk':'Unknown'})
df_na_replaced.head(10)["bhk"]
```

```python
#11. Fill missing values in specific column with the mean of the respective column. 
#Without inplace
df.fillna({'total_sqft':df['total_sqft'].mean()})
```

```python
#11.a. Fill missing values in numerical columns with the mean of the respective column. Inplace
#df.fillna({'total_sqft':df['total_sqft'].mean()},inplace=True)
numeric_columns = df.select_dtypes(include=['number']).columns
df[numeric_columns] = df[numeric_columns].apply(lambda col: col.fillna(col.mean()))
```

```python
#12. Sort the DataFrame based on the 'price' and 'price_per_sqft' column in ascending and descending order.
df.sort_values(by=['price','price_per_sqft'],ascending=[True,False])
```

```python
#13. Create a new column called 'bath_per_bhk' by dividing 'bath' by 'bhk'.
df['bath_per_bhk']=df['bath']/df['bhk']
```

```python
#14. Replace all values in the 'size' column where the value is '1 BHK' with '1 Bedroom'.
df['size'] = df['size'].replace({'1 BHK':'1 Bedroom'})
```

```python
#15. Create and Drop the 'availability' column from the dataset.
df['availability']='Y'
df.drop(columns=['availability'])
```

```python
#16. Select rows where the 'price' is greater than 1 crore (10 million).
df[df['price'] > 100]
```

```python
#17. Select rows where the 'location' is 'Whitefield'.
df[df['location'] == 'Whitefield']
```

```python
#18. Select rows where the 'size' is '2 BHK' and 'price' is less than 50 lakhs (5 million).
df[(df['size'] == '2 BHK') & (df['price'] < 50)]
```

```python
#19. Select rows where the 'location' contains the word 'Indira'.
df[df['location'].str.contains('indira', case=False)]
```

```python
#20. Select rows where the 'total_sqft' is missing.
df[df['total_sqft'].isna()]
```

```python
#21. Group the data by 'location' and calculate the average 'price' for each location.
df.groupby('location')['price'].mean()
```

```python
#22. Find the total number of houses available in each 'location'.
df['location'].value_counts()
```
